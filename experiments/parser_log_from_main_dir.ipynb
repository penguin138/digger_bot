```python


sentences  = []
uselstr = ["--------"]

chat_id   = 0
client_id = 0
chatstr   = 'Chat id = '
clientstr = 'Client id = '
client    = 'Клиент' 
employee  = 'Сотрудник'


for i in range(40):
    sent = f.readline()
    if i==0:
        sent = sent[1:]
    sent = sent[:-1]
    
    usel = False
    for word in uselstr:
        if word in sent:
            usel = True
    
    if sent == '':
        usel = True
        
    if chatstr in sent:
#        print(chat)
        chat_id = int(sent[len(chatstr):]) 
#        sentences.append(chat_id)
        usel = True

    if clientstr in sent:
#        print(sent)
        client_id = int(sent[len(clientstr):])    
#        sentences.append(client_id)
        usel = True
    
    if not usel:

        time = sent[:8]
        sent = sent[9:]
        if client in sent:
            sent = sent[8:]
            who = client
        if employee in sent:
            sent = sent[11:]
            who = employee
        
        sentences.append( (chat_id, client_id, time, who, sent) )
```
